# Agregación de datos por categoría

# Asi se crea un dummy dataframe:

```python
import numpy as np
import pandas as pd
```

```python
genero = ["Masculino", "Femenino"]
ingresos = ["Pobre", "Clase media", "Rico"]
```

```python
n =500
datos_genero = []
ingresos_datos = []

for i in range(0,500):
    datos_genero.append(np.random.choice(genero))
    ingresos_datos.append(np.random.choice(ingresos))
```

```python
datos_genero
```

```python
ingresos_datos
```

```python
#Z -> N (0,1)
#N(m,s) -> m+s * z
Altura = 160 + 30 * np.random.randn(n)
Peso = 65 + 30 * np.random.randn(n)
Edad = 30 + 12 * np.random.randn(n)
Ingresos = 18000 + 3500 * np.random.randn(n)
```

```python
datos = pd.DataFrame (
    {
        "Genero": datos_genero,
        "Status Economico": ingresos_datos,
        "Altura"  : Altura,
        "Peso"    : Peso,
        "Edad"    : Edad,
        'Ingresos': Ingresos
    }

)
```

```python
datos.head()
```

# Agrupación de datos

```python
genero_agrupado = datos.groupby("Genero")
```

```python
genero_agrupado.groups
```

```python
for names, groups in genero_agrupado:
    print(names)
    print(groups)
```

```python
genero_agrupado.get_group("Femenino")
```

```python
doble_agrupacion = datos.groupby(["Genero", "Status Economico"])
```

```python
len(doble_agrupacion)
```

```python
for names, groups in doble_agrupacion:
    print(names)
    print(groups)
```

# Operaciones sobre datos agrupados

```python
doble_agrupacion.sum()
```

```python
doble_agrupacion.describe()
```

```python
ingresos_agrupados = doble_agrupacion["Ingresos"]
```

```python
ingresos_agrupados.describe()
```

```python
doble_agrupacion.aggregate(
    {
        "Ingresos": np.sum,
        "Edad"    : np.mean,  
        "Altura"  : np.std
    }
)
```

```python
doble_agrupacion.aggregate(
    {
        "Edad": np.mean,
        "Altura": lambda h:np.mean(h)/np.std(h)
        
    }
)
```

```python
doble_agrupacion.aggregate([np.sum, np.mean, np.std])
```

```python
doble_agrupacion.aggregate([lambda x: np.mean(x) / np.std(x)])
```

# # Filtrado de datos

```python
doble_agrupacion["Edad"].filter(lambda x: x.sum()>2400)
```

# Transformacion de variables

```python
zscore = lambda x : (x- x.mean())/x.std()
```

```python
doble_agrupacion.transform(zscore)
```

```python
llenar_nas =lambda x: x.fillna(x.mean()) #Reenplaza los valores vacios (na) con el valor promedio
```

```python
doble_agrupacion.transform(llenar_nas)
```
