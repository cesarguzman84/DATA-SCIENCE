# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
datos = pd.read_csv(r"E:\VARIOS\python-ml-course-master\datasets\movies\movies.csv", sep=";")
datos
```

```python
peliculas = datos.columns.values.tolist()[1:]
peliculas
```

```python
matriz_distancias1 = distance_matrix(datos[peliculas], datos[peliculas], p=1)
matriz_distancias2 = distance_matrix(datos[peliculas], datos[peliculas], p=2)
matriz_distancias10 = distance_matrix(datos[peliculas], datos[peliculas], p=10)
```

```python
def matriz_distancia_data_frame(matriz_distancias, col_name):
    import pandas as pd
    return pd.DataFrame(matriz_distancias, index=col_name, columns=col_name)
```

```python
matriz_distancia_data_frame(matriz_distancias1, datos["user_id"])
```

```python
matriz_distancia_data_frame(matriz_distancias2, datos["user_id"])
```

```python
matriz_distancia_data_frame(matriz_distancias10, datos["user_id"])
```

# Clustering jerarquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
peliculas
```

```python
datos[peliculas]
```

```python
z = linkage(datos[peliculas], "ward")
z
```

```python
plt.figure(figsize = (25,10))
plt.title("Dendrograma jerrquico para el clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```
