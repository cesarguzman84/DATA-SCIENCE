# La probabilidad condicional

```python
from IPython.display import display, Math, Latex
```

* ¿Cuál es la probabilidad de que un cliente compre un producto sabiendo que es un hombre?

* ¿Cuál es la probabilidad de que sabiendo que un cliente compra un producto sea mujer?

```python
display(Math(r'P(Compra|Hombre )= \frac{Numero\ total\ de\ compras\ hechas\ por\ hombres}{Numero\ total\ de\ hombres\ del\ grupo} = \frac{Compra\cap Hombre}{Hombre}'))
```

```python
121/246
```

```python
display(Math(r'P(Numero\ Compra|Hombre) = 1-P(Compra|Hombre)'))
125/246
```

```python
display(Math(r'P(Mujer|Compra) = \frac{Numero\ total\ de\ compras\ hechas\ por\ mujeres}{Numero\ total\ de\ compras} = \frac{Mujer\cap Compra}{Compra}'))
159/280
```

```python
display(Math(r'P(Hombre|Compra)'))
121/280
```

```python
display(Math(r'P(Compra|Hombre)'))
print(124/246)
display(Math(r'P(NO\ Compra|Hombre)'))
print(125/246)
display(Math(r'P(Compra|Mujer)'))
print(159/265)
display(Math(r'P(NO\ Compra|Mujer)'))
print(106/265)
```

# Ratio de probabilidades

## Cociente entre los casos de exito sobre los de fracaso en el suceso estudiado y para cada grupo

```python
display(Math(r'P_h = probabilidad\ de\ hacer\ compra\ sabiendo\ que\ es\ un\ hombre'))
display(Math(r'P_m = probabilidad\ de\ hacer\ compra\ sabiendo\ que\ es\ una\ mujer'))
display(Math(r'odds_{compras, hombre}= \frac{P_h}{1-P_h} ')) 
display(Math(r'odds_{compras, mujer}= \frac{P_m}{1-P_m}'))
```

```python
ph=121/246
pm= 159/265
odds_h = ph/(1-ph) #121/125
odss_m = pm/(1-pm) #159/106
```

```python
odds_h
```

```python
odss_m
```

* Si el ratio es superior a 1, es mas probable el exito que el fracaso. Cuanto mayor es el ratio, mas probabilidad de exito de nuestro suceso
* Si el ratio es exactamente igual a 1, existo y fracaso son equiprobables (p=0.5)
* Si el ratio es menor que 1, el fracaso es mas probable que el exito. Cuanto menor es el ratio, menor probabilidad de exito de nuestro suceso
