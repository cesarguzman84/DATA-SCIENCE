# Funciones de distribución de probabilidades

# Distribución Uniforme

```python
#Distribucion uniforme: genera numeros aleatorios equiprobables (con la misma probabilidad)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# n es el numero de muestras entre a y b
a=1
b = 100
n = 1000000
datos = np.random.uniform(a, b, n)
```

```python
%matplotlib inline
plt.hist(datos)
```

# Distribución normal

```python
datos = np.random.randn(1000000)
```

```python
x = range(1, 1000001)
plt.plot(x, datos)
```

```python
plt.hist(datos)
```

```python
plt.plot(x,sorted(datos)) #Districunion de probabilidad acumulada
```

```python
media = 5.5
sd = 2.5 #desviacion estandar
datos = media + sd*np.random.randn(10000)
plt.hist(datos)
```
