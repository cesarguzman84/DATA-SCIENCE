# El tratamiento de las variables categóricas

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"E:\VARIOS\python-ml-course-master\datasets\ecom-expense\Ecom Expense.csv")
```

```python
df.head()
```

```python
dummy_genero = pd.get_dummies(df["Gender"], prefix = "Gender")
dummy_ciudad = pd.get_dummies(df["City Tier"], prefix="City")
```

```python
dummy_genero.head()
```

```python
dummy_ciudad.head()
```

```python
nombres_columnas = df.columns.values.tolist()
nombres_columnas
```

```python
nuevo_df = df [nombres_columnas].join(dummy_genero)
nombres_columnas = nuevo_df.columns.values.tolist()
nuevo_df.head()
```

```python
nuevo_df = nuevo_df [nombres_columnas].join(dummy_ciudad)
nuevo_df.head()
```

```python
caracteristica_columnas =["Monthly Income", "Transaction Time", 
                          "Gender_Female", "Gender_Male",
                         "City_Tier 1","City_Tier 2","City_Tier 3",
                         "Record"]
```

```python
X = nuevo_df [caracteristica_columnas]
Y = nuevo_df ["Total Spend"]
```

```python
modelo_lineal = LinearRegression()
modelo_lineal.fit(X,Y)
```

```python
print(modelo_lineal.intercept_)
print(modelo_lineal.coef_)
```

```python
list(zip(caracteristica_columnas, modelo_lineal.coef_))
```

```python
modelo_lineal.score(X,Y)
```

```python
SSD = np.sum((nuevo_df ["prediction"] - nuevo_df["Total Spend"])**2)
```

```python
SSD
```

```python
RSE = np.sqrt(SSD/(len(nuevo_df)-len(caracteristica_columnas)-1))
```

Otra posible forma de hacer el cálculo de las predicciones podría ser en usando la función predict del modelo lineal y añadiendolo como columna adicional al data frame:

# Transformación de variables para conseguir una relación no linel

```python
import pandas as pd
```

```python
datos_autos = pd.read_csv(r"E:\VARIOS\python-ml-course-master\datasets\auto\auto-mpg.csv")
```

```python
datos_autos.head()
```

```python
datos_autos.shape
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
datos_autos ["mpg"] =datos_autos["mpg"].dropna()
datos_autos ["horsepower"] = datos_autos ["horsepower"].dropna()
plt.plot(datos_autos["horsepower"], datos_autos ["mpg"], "ro")
plt.xlabel("Caballos de Potencia")
plt.ylabel("Consumo (millas por galón)")
plt.title("Caballos de fuerza vs MPG")
```

### Modelo de regresión lineal
* mpg = a + b * horsepower

```python
X = datos_autos ["horsepower"].fillna(datos_autos["horsepower"].mean())
Y = datos_autos ["mpg"].fillna(datos_autos["mpg"].mean())
X_datos = X [:, np.newaxis]
```

```python
modelo_lineal = LinearRegression()
modelo_lineal.fit(X_datos,Y)
```

```python
type(X)
```

```python
%matplotlib inline
plt.plot(X,Y, "ro")
plt.plot(X, modelo_lineal.predict(X_datos), color = "blue")
```

```python
modelo_lineal.score(X_datos, Y)
```

```python
SSD = np.sum((Y - modelo_lineal.predict(X_datos))**2)
RSE = np.sqrt(SSD/(len(X_datos)-1))
y_promedio =np.mean(Y)
error = RSE / y_promedio
SSD, RSE, y_promedio, error
```

### Modelo de regresión cuadrático 

* mpg = a+ b * horsepower^2

```python
X_datos = X**2
X_datos = X_datos [:, np.newaxis]
```

```python
modelo_lineal = LinearRegression()
modelo_lineal.fit(X_datos, Y)
```

```python
modelo_lineal.score(X_datos,Y)
```

```python
SSD = np.sum((Y - modelo_lineal.predict(X_datos))**2)
RSE = np.sqrt(SSD/(len(X_datos)-1))
y_promedio =np.mean(Y)
error = RSE / y_promedio
SSD, RSE, y_promedio, error
```

mpg = 55.026 - 0.434 hp + 0.00112615 hp^2

```python
for d in range (2,6):
    poly = PolynomialFeatures(degree=d)
    X_datos = poly.fit_transform(X[:,np.newaxis])
    modelo_lineal = linear_model.LinearRegression()
    modelo_lineal.fit(X_datos, Y)
    print("Regresión de grado " + str(d))
    print(modelo_lineal.score(X_datos, Y))
    validacion_regresion(X_datos, Y, modelo_lineal.predict(X_datos))
```

```python
def validacion_regresion (X_datos,Y, Y_pred):
    SSD = np.sum((Y - Y_pred)**2)
    RSE = np.sqrt(SSD/(len(X_datos)-1))
    y_promedio =np.mean(Y)
    error = RSE / y_promedio
    print("SSD: " + str(SSD) + ", RSE: " +str(RSE) + ", y_promedio: " + str(y_promedio) + ", error: " + str(error*100)+"%")
```
