# RegresiÃ³n logistica con el paquete statsmodel de python

```python
import numpy as np
```

```python
X = np.array(range(10)).reshape(10,1)
X
```

```python
Y = [0,0,0,0,1,0,1,0,1,1]
```

```python
bias = np.ones(10).reshape(10,1)
X_nuevo = np.append(X, bias, axis =1)
```

```python
import statsmodels.api as sm
```

```python
modelo_logistico = sm.Logit(Y, X_nuevo)
```

```python
resultado = modelo_logistico.fit()
print(resultado.summary2())
```
