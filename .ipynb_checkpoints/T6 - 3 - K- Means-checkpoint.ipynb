# El m√©todo de k-means

```python
import numpy as np
```

```python
datos = np.random.random(90).reshape(30,3)
datos
```

```python
c1 = np.random.choice(range(len(datos)))
c2 = np.random.choice(range(len(datos)))
centro_clusters = np.vstack([datos[c1], datos[c2]])
centro_clusters
```

```python
from scipy.cluster.vq import vq
```

```python
vq(datos, centro_clusters)
```

```python
from scipy.cluster.vq import kmeans
```

```python
kmeans(datos, centro_clusters)
```

```python
kmeans(datos, 2)
```
