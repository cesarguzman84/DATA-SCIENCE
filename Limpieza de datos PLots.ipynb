# Plots y visualizacion de los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv(r"E:\VARIOS\python-ml-course-master\datasets\customer-churn-model\Customer Churn Model.txt")
```

```python
datos.head()
```

```python
%matplotlib inline   
```

# Scatter Plot

```python
datos.plot(kind = "scatter", x="Day Mins", y="Day Charge")
```

```python
figura, ejes = plt.subplots(2,2, sharey=True, sharex=True)
datos.plot(kind="scatter", x="Day Mins", y="Day Charge", ax=ejes[0][0])
datos.plot(kind="scatter", x="Night Mins", y="Night Charge", ax=ejes[0][1])
datos.plot(kind="scatter", x="Day Calls", y="Day Charge", ax=ejes[1][0])
datos.plot(kind="scatter", x="Night Calls", y="Night Charge", ax=ejes[1][1])
```

# Histograma de frecuencias 

```python
plt.hist(datos["Day Calls"], bins=20)
plt.xlabel("Numero de llamadas al dia")
plt.ylabel("Frecuencia")
plt.title("Histograma de numero de llamadas al dia")
```

# Boxplot, diagrama de caja y bigotes

```python
plt.boxplot(datos["Day Calls"])
plt.ylabel("Numero de llamadas diarias")
plt.title("Boxplot de las llamadas diarias")
```
